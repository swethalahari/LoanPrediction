```python
def get_combined_data():
    train = pd.read_csv('train_Loan.csv')
    test = pd.read_csv('test_Loan.csv')
    targets = train.Loan_Status
    train.drop('Loan_Status', 1, inplace=True)
    combined_ds = train.append(test)
    combined_ds.reset_index(inplace=True)
    combined_ds.drop(['index', 'Loan_ID'], inplace=True, axis=1)
    return combined_ds
```

```python
combined_ds = get_combined_data()
combined_ds.describe()
```

```python
combined_ds.isnull().sum()
```

```python
combined_ds['Gender'].fillna('Male', inplace=True)
```

```python
combined_ds['Married'].fillna('Yes', inplace=True)
```

```python
combined_ds['Self_Employed'].fillna('No', inplace=True)
```

```python
combined_ds['LoanAmount'].fillna(combined_ds['LoanAmount'].median(), inplace=True)
```

```python
combined_ds['Credit_History'].fillna(2, inplace=True)
```

```python
numerics = ['float64', 'int64']
objects_ds = combined_ds.select_dtypes(exclude=numerics)
```

```python
newdatadf = objects_ds
for i in list(objects_ds.columns):
    newdatadf[i] = pd.Categorical(objects_ds[i]) # Convert all the variables into categorical
dummy = pd.get_dummies(newdatadf)
dummy.describe
```

```python
combined_ds['Gender'] = combined_ds['Gender'].map({'Male':1,'Female':0})
```

```python
combined_ds['Education'] = combined_ds['Education'].map({'Graduate':1,'Not Graduate':0})
```

```python
combined_ds['Self_Employed'] = combined_ds['Self_Employed'].map({'Yes':1,'No':0})
```

```python
combined_ds['Total_income'] = combined_ds['ApplicantIncome']+combined_ds['CoapplicantIncome']
combined_ds.drop(['ApplicantIncome','CoapplicantIncome'],inplace=True,axis = 1)
```

```python
 pd.get_dummies(combined_ds, columns=['Dependents'])
```

```python
combined_ds['Debt_Income_Ratio'] = combined_ds['Total_income'] / combined_ds['LoanAmount']
```

```python
combined_ds['Loan_Amount_Term'].value_counts()
```

```python
def process_Loan_Amt_term():
    global combined_ds
    combined_ds['Very_Short_Term'] = combined_ds['Loan_Amount_Term'].map(lambda t: 1 if t<=60 else 0)
    combined_ds['Short_Term'] = combined_ds['Loan_Amount_Term'].map(lambda t: 1 if t>60 and t<180 else 0)
    combined_ds['Long_Term'] = combined_ds['Loan_Amount_Term'].map(lambda t: 1 if t>=180 and t<=300  else 0)
    combined_ds['Very_Long_Term'] = combined_ds['Loan_Amount_Term'].map(lambda t: 1 if t>300 else 0)
    combined_ds.drop('Loan_Amount_Term', axis=1, inplace=True)
```

```python
process_Loan_Amt_term()
```

```python
def process_credit_history():
    global combined_ds
    combined_ds['Credit_History_Bad'] = combined_ds['Credit_History'].map(lambda c: 1 if c==0 else 0)
    combined_ds['Credit_History_Good'] = combined_ds['Credit_History'].map(lambda c: 1 if c==1 else 0)
    combined_ds['Credit_History_Unknown'] = combined_ds['Credit_History'].map(lambda c: 1 if c==2 else 0)
    combined_ds.drop('Credit_History', axis=1, inplace=True)
```

```python
process_credit_history()
```

```python
def process_property():
    global combined_ds
    property_dummies = pd.get_dummies(combined_ds['Property_Area'], prefix='Property')
    combined_ds = pd.concat([combined_ds, property_dummies], axis=1)
    combined_ds.drop('Property_Area', axis=1, inplace=True)
```

```python
process_property()
```

```python
def feature_scaling(df):
    df -= df.min()
    df /= df.max()
    return df
```

```python
combined_ds['LoanAmount'] = feature_scaling(combined_ds['LoanAmount'])
combined_ds['Total_Income'] = feature_scaling(combined_ds['Total_income'])
combined_ds['Debt_Income_Ratio'] = feature_scaling(combined_ds['Debt_Income_Ratio'])
```

```python
combined_ds[100:110]
```

```python
combined_ds['Married'] = combined_ds['Married'].map({'Yes':1,'No':0})
```

```python
def process_dependents():
    combined_ds['Singleton'] = combined_ds['Dependents'].map(lambda d: 1 if d=='1' else 0)
    combined_ds['Small_Family'] = combined_ds['Dependents'].map(lambda d: 1 if d=='2' else 0)
    combined_ds['Large_Family'] = combined_ds['Dependents'].map(lambda d: 1 if d=='3+' else 0)
    combined_ds.drop(['Dependents'], axis=1, inplace=True)
    
```

```python
def recover_train_test_target():
    targets = data_train['Loan_Status'].map({'Y':1,'N':0})
    train = combined_ds.head(614)
    test = combined_ds.iloc[614:]
    return train, test, targets
```

```python
train, test, targets = recover_train_test_target()
```
